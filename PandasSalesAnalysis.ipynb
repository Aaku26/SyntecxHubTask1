```python
import pandas as pd

# Load the dataset and calculate sales per product
df = pd.read_excel('superstore_sales.xlsx')
prod_sales = pd.DataFrame(df.groupby('product_name').sum(numeric_only=True)['sales'])
```

```python
# Top 10
prod_sales.sort_values('sales',ascending=False)
```

```python
prod_sales[:10]
```

**Most Selling products**

```python
#Grouping product name
most_sell_prod = pd.DataFrame(df.groupby('product_name').sum(numeric_only=True)['quantity'])
```

```python
#Sorting most_sell_products
most_sell_prod = most_sell_prod.sort_values('quantity',ascending=False)
```

**Most Preferred ship mode**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Setting figure size
plt.figure(figsize=(10, 8.5))

# Plotting ship mode count vertically
sns.countplot(x=df['ship_mode'])
plt.title('Count of Orders by Ship Mode')
plt.show()
```

**Most Profitable Category and Sub-category**

```python
#Grouping category and subcategory
cat_subcat_profit = pd.DataFrame(df.groupby(['category', 'sub_category']).sum(numeric_only=True)['profit'])
```

```python
cat_subcat_profit.sort_values('profit',ascending=False)
```
